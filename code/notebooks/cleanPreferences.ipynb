```
#Import essential libraries
import pandas as pd
from google.colab import files
```

```
#Load preferences data
#Data Source - https://www.kaggle.com/steventaylor11/stated-preferences-for-car-choice
preferences = pd.read_csv("preferences.csv")
preferences.head()
```

```
#Create dictionaries for cleaning attribute names
binaryDict = {0: "N", 1: "Y"}
typeDict = {"regcar": "regular car", "sportuv": "sport utility vehicle", "sportcar": "sport car",
            "stwagon": "station wagon", "truck": "truck", "van": "van"}
fuelDict = {"cng": "compressed natural gas", "gasoline": "gasoline", "methanol": "methanol", "electric": "electric"}
sizeDict = {0: "mini", 1: "subcompact", 2: "compact", 3: "mid-size/large"}


#Variable descriptions
"""
Y: Yes
N: No
Price: price of vehicle divided by the logarithm of income
Speed: highest attainable speed (mph)
Pollution: tailpipe emissions as fraction of those for new gas vehicle
"""

# Create new preferences dictionary
newDict = {"College Education": [], "Household Size > 2": [], "Daily Commute < 5 miles": [],
                 "Type": [], "Fuel": [], "Price": [], "Speed": [],
                 "Pollution": [], "Size": []}


#Iterate through each row
for index in preferences.index:
  row = preferences.iloc[index]
  choice = row["choice"][-1] #choice selected by individual

  #Add data to newDict dictionary
  newDict["College Education"].append(binaryDict[row["college"]])
  newDict["Household Size > 2"].append(binaryDict[row["hsg2"]])
  newDict["Daily Commute < 5 miles"].append(binaryDict[row["coml5"]])

  newDict["Type"].append(typeDict[row["type" + choice]])
  newDict["Fuel"].append(fuelDict[row["fuel" + choice]])
  newDict["Size"].append(sizeDict[row["size" + choice]])

  newDict["Price"].append(row["price" + choice])
  newDict["Speed"].append(row["speed" + choice])
  newDict["Pollution"].append(row["pollution" + choice])
```

```
#Create new preferences data frame from newDict dictionary
preferences = pd.DataFrame(newDict)
preferences.head()
```

```
#Download new preferences data
preferences.to_csv("preferencesData.csv")
files.download("preferencesData.csv")
```
